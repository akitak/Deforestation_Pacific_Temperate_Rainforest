```python
# area_burn_Ha, area_tree_cover_loss_Ha

plt.figure(figsize=[20, 12])

plt.bar(central_coast.index, central_coast['area_burn_Ha'], color='red', label = 'Area burned (Ha)', width = 0.25)
plt.bar(central_coast.index + 0.25, central_coast['area_tree_cover_loss_Ha'], color='blue', label = 'Area lost (Ha)', width = 0.25)
plt.bar(central_coast.index + 0.5, 0, width = 0.25)
plt.legend()

plt.xticks([i + 0.25 for i in range(2001, 2021)], central_coast.index)

plt.title("Forest area lost due to Tree Cover Loss and Fire")

plt.xlabel('Year')
plt.ylabel('Area lost (Ha)')

plt.show();
```

```python
central_coast['cumulative_loss_area'] = central_coast['area_tree_cover_loss_Ha'] +  central_coast['area_burn_Ha']
```

```python
series_forest_area_years = [value for value in central_coast.index]
```

```python
series_forest_area_years
```

```python
current_area = 1.603821e+06
series_forest_area_year_start = [1.603821e+06]
for year in series_forest_area_years[:-1]:
    current_area = current_area - float(central_coast[central_coast.index == year]['cumulative_loss_area'])
    series_forest_area_year_start.append(current_area)
```

```python
series_forest_area_year_start = np.array(series_forest_area_year_start)
```

```python
central_coast_df = central_coast.copy()
central_coast_df['true_area_tree_cover_jan1'] = series_forest_area_year_start
central_coast_df['true_area_tree_cover_dec31'] = central_coast_df['area_tree_cover_jan1'] - central_coast_df['cumulative_loss_area']
central_coast_df['true_percentage_tree_cover_loss'] = central_coast_df['cumulative_loss_area'] / central_coast_df['true_area_tree_cover_jan1']
```

```python
central_coast_df
```

```python
plt.figure(figsize = (12, 12))
plt.plot(central_coast_df.index, central_coast_df['percentage_tree_cover_loss'], color = 'blue', label = 'Percentage of Tree Cover Loss')
plt.plot(central_coast_df.index, central_coast_df['percentage_tree_cover_loss_over_logging_land'], color = 'black', label = 'Percentage of Tree Cover Loss compared to Logging Approved Area')
plt.plot(central_coast_df.index, central_coast_df['true_percentage_tree_cover_loss'], color = 'red', label = 'Percentage of Tree Cover Loss including Fire')
plt.axhline(y=0.02/100, color='r', linestyle='-')
plt.title('Deforestation Rate over Time')
plt.xticks(central_coast_df.index)
plt.xlabel('Year')
plt.ylabel('Percentage of Area loss')
plt.legend();
```

```python
plt.figure(figsize = (12, 12))
sns.scatterplot(central_coast_df['TMAX'], central_coast_df['true_percentage_tree_cover_loss'], color = 'blue', label = 'Percentage of Tree Cover Loss including Fire vs TMIN')
sns.scatterplot(central_coast_df['TAVG'], central_coast_df['true_percentage_tree_cover_loss'], color = 'red', label = 'Percentage of Tree Cover Loss including Fire vs TAVG')
plt.axhline(y=0.02/100, color='r', linestyle='-')
plt.title('Deforestation Rate vs Temperature')
plt.xlabel('Temperature (Celsius)')
plt.ylabel('Percentage of Area loss')
plt.legend();
```
